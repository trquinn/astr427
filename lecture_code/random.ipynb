```python
from scipy.stats import qmc
```

```python
sampler = qmc.Sobol(d=2)
sample = sampler.random_base2(10)
```

```python
plt.scatter(sample[:,0],sample[:,1])
```

```python
sampler2 = qmc.Sobol(d=2, optimization="lloyd")
```

```python
sample2 = sampler2.random_base2(10)
```

```python
plt.scatter(sample2[:,0],sample2[:,1])
```
