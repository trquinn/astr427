```python
import numpy as np

def f(x) :
    global xsample
    xsample.append(x[0])
    return x**2 + 2*np.sin(10*x)
```

```python
xint = np.linspace(-5, 5, 200)
plt.plot(xint, f(xint))
```

```python
import scipy.optimize as opt
```

```python
xsample=[]
soln = opt.dual_annealing(f, [(-5.0, 5.0)], maxiter = 100, initial_temp=25, no_local_search=True)
```

```python
print(soln)
```
