```python
import numpy as np
import scipy.integrate as integ
```

```python
# Example Damped Driven harmonic oscilator
# x'' + B x' + C = D cos(Omega t)
def f(t, y) :
    # t is the independent variable; y is the array of dependent variables
    # where I identify y[0] as "x" in the above equation and
    # y[1] is identified as x' \equiv v
    B = 1.0
    C = 1.0
    D = 1.0
    Omega = 5.0
    dydt = np.array([0.0,0.0])
    dydt[0] = y[1]  # dx/dt = v
    dydt[1] = -B*y[1] - C*y[0] + D*np.cos(Omega*t)
    return dydt
```

```python
# initial conditions: stationary and displaced from 0
y_0 = np.array([1.0, 0.0])
```

```python
t_eval = np.arange(0.0, 10.0, 0.1)
```

```python
soln = integ.solve_ivp(f, (0.0,10.), y_0, t_eval=t_eval)
```

```python
import pylab as plt
plt.plot(soln.t, soln.y[0])
```

```python
plt.plot(soln.t, soln.y[1])
```

```python
# An example of using solve_ivp() and forcing it to used a fixed
# timestep.
soln = integ.solve_ivp(f, (0.0,10.), y_0, atol=1.0, rtol=1.0,first_step=.25,max_step=.25)
```

```python
import pylab as plt
plt.plot(soln.t, soln.y[0])
```
