```python
import numba as nb
import numpy as np
nb.get_num_threads()
```

```python
@nb.njit(parallel=True)
def mc_pi(nsamp) :
    total = 0
    delta = 1000000
    count = 0
    nloops = nsamp/delta
    for i in nb.prange(int(nloops)) :
        nextsamp = delta
        sampx = np.random.random(nextsamp)
        sampy = np.random.random(nextsamp)
        accept = (sampx**2 + sampy**2) < 1.0
        #Warning: critical section!
        count += np.count_nonzero(accept)
        total += delta
        
    return 4*count/nsamp
```

```python
%time mc_pi(1000000000)
```

```python
# Broken example
@nb.njit(parallel=True)
def mc_pi(nsamp) :
    total = 0
    delta = 1000000
    count = np.zeros(2)
    nloops = nsamp/delta
    for i in nb.prange(int(nloops)) :
        nextsamp = delta
        sampx = np.random.random(nextsamp)
        sampy = np.random.random(nextsamp)
        accept = (sampx**2 + sampy**2) < 1.0
        #Warning: critical section!
        count[i%2] += np.count_nonzero(accept)
        total += delta
        
    return 4*count.sum()/nsamp
```

```python
%time mc_pi(1000000000)
```
