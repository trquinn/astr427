```python
import numpy as np
from  scipy.interpolate import lagrange
from  scipy.interpolate import interp1d

x = np.array([0.0, 1.4, 1.6, 3.0])
y = np.array([1.0, 1.0, -1.0, -1.0])
poly = lagrange(x, y)
xint = np.arange(0.0, 3.1, 0.1)
yintp = poly(xint)
#spline = interp1d(x, y, 'quadratic')
# yints = spline(xint)
```

```python

import pylab
pylab.plot(x, y, marker="x", label="data")
pylab.plot(xint, yintp, label="Lagrange")
# pylab.plot(xint, yints, label="spline")
pylab.legend()

```
